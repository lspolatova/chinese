```python
import re
letters = set()
import codecs
with codecs.open('chinese-word-list.txt', encoding = 'utf-8') as f:
    for line in f:
        line = re.sub("^\s+|\n|\r|0|3|5|\s+$", '', line)
        for chin in re.findall(r'[\u3400-\u4BDF\u4E00-\u9FFF\uF900-\uFAFF\u3400-\u4BDF\u4E00-\u9FFF\uF900-\uFAFF]', line):
            for letter in line:
                letters.add(letter)
print(letters)
f.closed
```
