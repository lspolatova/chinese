```python
import re
letters = set()
import codecs
with codecs.open('chinese-word-list.txt', encoding = 'utf_8') as f:
    count = 0
    for line in f:
        count = count + 1
        line = re.sub("^\s+|\n|\r|\s+$", '', line)
        for letter in line:
            letters.add(letter)
        if(not(count%60000)):
            print(letters)
            break
#print(letters)
f.closed
```
